```python
df5=pd.read_csv('merged_data.csv')
df5.head()
```

```python
print("Existing columns:", df5.columns)
```

```python
df5[['Date', 'Time', 'DS18B20_Temperature', 'SN1_WE', 'SN1_Aux', 'NO2_Concentration']] = df5['Date;Time;DS18B20_Temperature;SN1_WE;SN1_Aux;NO2_Concentration'].str.split(';', expand=True)

df5 = df5.drop('Date;Time;DS18B20_Temperature;SN1_WE;SN1_Aux;NO2_Concentration', axis=1)
print(df5)
```

```python
df5 = df5[['NO2_Concentration']]
print(df5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error
X = df5[['NO2_Concentration']]
Y_original = df5['NO2_Concentration']

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y_original, test_size=0.2, random_state=42)

# Linear Regression
linear_model = LinearRegression()
linear_model.fit(X_train, Y_train)

# Predict Y values using the trained model
Y_predicted = linear_model.predict(X_test)
print(Y_predicted)
```

```python
mlp_regression = MLPRegressor(hidden_layer_sizes=(100, 50), max_iter=1000, random_state=42)
mlp_regression.fit(X_train, Y_train)
mlp_ypred = mlp_regression.predict(X_test)
print(mlp_ypred)
```

```python
logistic_regression = LogisticRegression()
logistic_regression.fit(X_train, Y_train)
logistic_ypred = logistic_regression.predict(X_test)
print(logistic_ypred)
```
